```python
df = pd.read_sql_query('SELECT BookID,BookTitle,BookAuthor,BookGenre FROM Book',conn)
```

```python
print(df)
```

```python
df = df.drop_duplicates(subset=['BookTitle'])
```

```python
df = df.dropna(axis=0)
```

```python
print(df.head())
```

```python
def clean_text(author):
    result = str(author).lower()
    return(result.replace(' ',''))
```

```python
df['BookAuthor'] = df['BookAuthor'].apply(clean_text)
```

```python
df
```

```python
df['BookTitle'] = df['BookTitle'].str.lower()
```

```python
df
```

```python
df2 = df
```

```python
df2['data'] = df2[df2.columns[1:]].apply(
    lambda x: ' '.join(x.dropna().astype(str)),
    axis=1
)
```

```python
df2
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
vectorized = vectorizer.fit_transform(df2['data'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity(vectorized)
```

```python
print(similarities)
```

```python
df3 = pd.DataFrame(similarities, columns=df2['BookTitle'], index=df2['BookID']).reset_index()
df3[:15]
```

```python
# pd.DataFrame(df3.nlargest(1,'ice monster')['BookTitle'])
df3[['BookID','']]
```

```python
input_book = 'the last olympian'
recommendations = pd.DataFrame(df3.nlargest(11,input_book)['BookID'])
recommendations = recommendations[recommendations['BookID']!=input_book]
print(recommendations)
```
