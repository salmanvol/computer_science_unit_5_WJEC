```python
matrix = df.pivot_table(index='StudentID', columns='BookTitle', values='borrowed')
```

```python
matrix
```

```python
# Similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
user_similarity = cosine_similarity(matrix.fillna(0))
user_similarity
```

```python
user_similarity = pd.DataFrame(user_similarity, columns = matrix.index.tolist())
```

```python
try:
    user_similarity = user_similarity.drop(index=0)
except:
    pass
```

```python
user_similarity
```

```python
# Pick a user ID
picked_userid = 1
# Remove picked user ID from the candidate list
try:
    user_similarity.drop(index=picked_userid, inplace=True)
except:
    print('picked_userid has already been removed')
# Take a look at the data
user_similarity.head()
```

```python
# Number of similar users
n = 10
# User similarity threashold
user_similarity_threshold = 0.3
# Get top n similar users
similar_users = user_similarity[user_similarity[picked_userid]>user_similarity_threshold][picked_userid].sort_values(ascending=False)[:n]

# Print out top n similar users
print(f'The similar users for user {picked_userid} are\n', similar_users)
```

```python
# Books that the target user has watched
picked_userid_borrowed = matrix[matrix.index == picked_userid].dropna(axis=1, how='all')
picked_userid_borrowed
```

```python
# Books that similar users watched. Remove movies that none of the similar users have watched
similar_user_books = matrix[matrix.index.isin(similar_users.index)].dropna(axis=1, how='all')
similar_user_books
```

```python
# Remove the watched movie from the movie list
similar_user_books.drop(picked_userid_borrowed.columns,axis=1, inplace=True, errors='ignore')
# Take a look at the data
similar_user_books
```

```python
item_score = {}

# Loop through items
for i in similar_user_books.columns:
    # Get the ratings for movie i
    book_rating = similar_user_books[i]
    # create a variable to store the score
    total = 0
    # create a variable to store the number of scores
    count = 0
    # Loop through similar users
    for u in similar_users.index:
        # if the movie has rating
        if not pd.isna(book_rating[u]):
            # Score is the sum of user similarity score mutiply by the movie rating
            score = similar_users[u] * book_rating[u]
            # Add the score to the total score for the movie so far
            total += score
            # Add 1 to the count
            count += 1
            
    item_score[i] = total / count
    
# Convert dictionary to pandas dataframe
item_score = pd.DataFrame(item_score.items(), columns=['book', 'book_score'])
    
# Sort the movies by score
ranked_item_score = item_score.sort_values(by='book_score', ascending=False)
# Select top m movies
m = 10
ranked_item_score.head(m)
```
